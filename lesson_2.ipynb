# 2. tách kiểu dữ liệu chuỗi

# 3. tách cột dữ liệu ngày tháng

```python
days = pd.to_datetime(df_center['date_collected'], format='%Y-%m-%d').dt.day
months = pd.to_datetime(df_center['date_collected'], format='%Y-%m-%d').dt.month
years  = pd.to_datetime(df_center['date_collected'], format='%Y-%m-%d').dt.year
df_center.insert(loc = 3, column='days', value = days)
df_center.insert(loc = 4, column = 'months', value = months)
df_center.insert(loc = 4, column = 'years', value = years)
```

```python
df_center.head()
```

# 4. tách cột dữ liệu thời gian

```python
df_center.info()
```

```python
hours = pd.to_datetime(df_center['response_time'], format = ' %H:%M:%S').dt.hour
minutes = pd.to_datetime(df_center['response_time'], format = ' %H:%M:%S').dt.minute
seconds = pd.to_datetime(df_center['response_time'], format = ' %H:%M:%S').dt.second
df_center.insert(loc = 13, column='second', value = seconds)
df_center.insert(loc = 14, column = 'minute', value = minutes)
df_center.insert(loc = 15, column = 'hour', value = hours)
```
